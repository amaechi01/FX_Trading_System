```python
import pandas as pd # type: ignore
import sys
sys.path.append('../')
import technicals.indicators as tech
from plotting import CandlePlot
```

```python
df = pd.read_pickle('../data/AUD_CAD_H1.pkl')
df.head()
```

```python
df_an = tech.BollingerBands(df)
df_an.tail()
```

```python
cp = CandlePlot(df_an[-200:], candles=True)
cp.show_plot(line_traces=['BB_MA', 'BB_UP', 'BB_LW'])
```

```python
df_atr = tech.ATR(df, n=14)
df_atr.tail()
```

```python
cp = CandlePlot(df_atr[-200:], candles=True)
cp.show_plot(sec_traces=['ATR_14'])
```

```python
df_kc = tech.KeltnerChannel(df)
df_kc.tail()
```

```python
cp = CandlePlot(df_kc.iloc[-200:], candles=True)
cp.show_plot(line_traces=['EMA', 'KeUp', 'KeLo'])
```

```python
df_rsi = tech.RSI(df)
df_rsi.tail()
```

```python
cp = CandlePlot(df_rsi[-100:], candles=True)
cp.show_plot(sec_traces=['RSI_14'])
```

```python
df_macd = tech.MACD(df)
df_macd.tail()
```

```python
cp = CandlePlot(df_macd.iloc[-100:], candles=True)
cp.show_plot()
```
