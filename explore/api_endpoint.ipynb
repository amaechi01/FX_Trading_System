Below, i will write a function that fetched candlesticks. This function will take the pair name, the number of the candlesticks and the granularity.
Among all the information that are returned as have been seen in the beginning, i would like my function to return the candles values which holds information about the different candlesticks

```python
def fetch_candles(pair_name, count=10, granularity='H1'):
    url = f'{crid.oanda_url}/instruments/{pair_name}/candles'
    params =dict(
        count = count,
        granularity = granularity,
        price = 'MBA'
    )
    response = session.get(url, params=params, data=None, headers=None)
    data = response.json()
    
    # the condition here is straight forward. if the status code is 200,
    #  check if the object contains the candles key. if the key is not there, return empty list
    # else return the candles values.
    #  if the status code is not 200, return the status code and the data
    if response.status_code == 200:
        if 'candles' not in data:
            data =[]
        else:
            data = data['candles']
    return response.status_code, data
```

Now i write a functio that converts the data to into a dataframe. I am interested in candlesticts that are completed.

```python
def get_candle_df(data):
    if len(data) == 0: # in case the returned data is empty, the function should return an empty dataframe should be etrurned.
        return pd.DataFrame()   
    prices = ['mid', 'bid', 'ask'] # to flatten the prices, i will over them and 
    ohlc = ['o', 'h', 'l', 'c'] # extract thr ohlc indivdiually
    final_data = [] # each of the candlesticks properties are appended in this list
    for item in data:
        if item['complete'] == False:
            continue
        dict_obj = {}
        dict_obj['time'] = parser.parse(item['time'])
        dict_obj['volume'] = item['volume']
        for price in prices:
            for _ in ohlc:
                dict_obj[f'{price}_{_}'] = float(item[f'{price}'][f'{_}'])
        final_data.append(dict_obj)
    return pd.DataFrame(final_data)

```

The final function in this section will fetch candlesticks and convert them to dataframes and save them as a picklefile

```python
def create_data_file(pair_name, count=10, granularity='H1'):
    code, data = fetch_candles(pair_name, count, granularity)
    if code != 200:
        print('failed', pair_name, data)
        return
    if len(data) == 0:
        print('No Candles', pair_name, data)
    candle_df = get_candle_df(data)
    candle_df.to_pickle(f'./data/{pair_name}_{granularity}.pkl')
    print(f'{pair_name} {granularity} {candle_df.shape[0]} candles, {candle_df.time.min()} {candle_df.time.max()}')
```

```python
create_data_file('EUR_USD', count=10, granularity='H4')
```
