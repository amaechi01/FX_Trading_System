**Detecting Trades**:
We shall narrow it down to just ma_50 and ma_200. The idea of the stragtegy is that is the smaller ma crosses above the longer ma, the place a buy order. If the smaller ma crosses below the longer ma value, we place a sell order. In every other case we should have no trade. So, when there is a buy order we have 1, -1 for a sell order and 0 where there was no trade

```python
# make variables of the ma values and buy, sell and no trade values as well
MA_S = 'MA_50'
MA_L = 'MA_200'
BUY = 1
SELL = -1
NONE = 0
```

```python
# filter the dataframe by the variables above and make a copy of it
df_an = df_ma[['time', 'mid_o', 'mid_h', 'mid_l', 'mid_c', MA_S, MA_L]].copy()
df_an
```

Below, we introduce a column - Delta which holds the difference between the short ma and the long ma

```python
df_an['Delta'] = df_an['MA_50'] - df_an['MA_200']
df_an.head()
```

To dictect a trade, we shall compare the signs values in the Delta columns to sign of the previous values. If there is no previous value, then there was no trade then move to the next one. OHow the signs interchange will determine the kind of trade that took place. if the signss remain the same, it implies that was no trade. To be able to compare the signs, we shall need to introduce another columns which will hold the forward shifted values of the Delta.

```python
df_an['Delta_Prev'] = df_an['Delta'].shift(1)
df_an.head()
```

```python
# is_trade will take a dataframe (in this case, our dataframe) and check if the 
#values of delta is greater than or equal to zero and the Delta_Prev is less 
# than zero to detect a buy order and in the opposite case to detect a sell order.
# if either of the two are not met, we have no order.
def is_trade(row):
    if row['Delta'] >= 0 and row['Delta_Prev'] < 0:
        return BUY
    elif row['Delta'] < 0 and row['Delta_Prev'] >= 0:
        return SELL
    return NONE
```

```python
# we apply is_trade to ur dataframe along axis 1
df_an['Trade'] = df_an.apply(is_trade, axis=1)
df_an.head()
```

```python
# we filter the dataframe to exclude the rows without trades
# check the shape know how many trades were placed
df_trades = df_an[df_an['Trade'] != 0].copy()
df_trades.shape
```

**STRATEGY ASSESSMENT**
we shall now use the infrastruture module to assees the performance of the strategy by leveraging on the pipLocation value. This will be used to calculate the gain or loss in pips of the given instrument

```python
ic.LoadInstruments('../data')
```

```python
ic.instruments_dict[pair]
```

```python
ins_data = ic.instruments_dict[pair]
```

```python
df_trades['Diff'] = df_trades['mid_c'].diff().shift(-1)
df_trades.fillna(0,inplace=True)
df_trades['Gain'] = df_trades['Diff']/ins_data.pipLocation
#df_trades['Gain'] = df_trades['Gain'] * df_trades['Trade']
df_trades.head()
```

When the pd.div() is applied to fetch the difference betwen the entry mid_c price and exit mid_c price, the difference appears to to be in the next row. to overcome this, the Diff column is shifted backwards.
The gain in pip will have to be the difference in entry mid_c and exit mid_c price divided by the pipLocation. Another issue is that the in cases were the entry price is lower than the closing price which should be a loss, we have positive pips. So, to tackle this, we multiply the gain by thre trade.

```python
#sum the gain column to check the overall performance
df_trades['Gain'].sum()
```

```python
# introduce a new column to hold the cumulative sum of the pips of each trade
df_trades['Gain_C'] = df_trades['Gain'].cumsum()
```

```python
# plot the cumulative sum of the performance of the strategy
cp = plot.CandlePlot(df_trades, candles=False)
cp.show_plot(line_traces=['Gain_C'])
```
